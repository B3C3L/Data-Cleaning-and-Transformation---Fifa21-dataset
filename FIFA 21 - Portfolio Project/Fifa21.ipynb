Converting lbs to kg

```python
def w_conv(x):
    x = x.replace("lbs","")
    a = str(round(int(x)*0.453592))
    return a + 'kg'
fifa['Weight'] = fifa['Weight'].map(w_conv)
```

```python
fifa['Weight']
```

Converting feet to cm

```python
def h_conv(x):
    x = x.replace("'", " ").replace("\"", " ").split()
    a = int(x[0]) * 30.48 + int(x[1]) * 2.54
    end = str(round(a))
    return  end + 'cm'
```

```python
fifa['Height']=fifa['Height'].map(h_conv)
fifa['Height']
```

Transforming Date to Datetime and separating column into year, month and day columns

```python
fifa.Joined = pd.to_datetime(fifa.Joined)
```

```python
fifa.info()
```

```python
fifa['Joined_Y'] = fifa['Joined'].dt.year
fifa['Joined_M'] = fifa['Joined'].dt.month
fifa['Joined_D'] = fifa['Joined'].dt.day
```

```python
fifa.columns
```

```python
fifa[['Value', 'Wage', 'Release Clause']]
```

Standardizing money values

```python
def standardize(x):
    if isinstance(x, str):
        if x.find('M') != -1:
            return int(float(x.replace('M','').replace('€', '')) * 1000000)
        elif x.find('K') != -1:
            return int(float(x.replace('K', '').replace('€', '')) * 1000)
    else:
        return int(x)
```

```python
fifa['Value'] = fifa['Value'].map(standardize)
fifa['Wage'] = fifa['Wage'].map(standardize)
fifa['Release Clause'] = fifa['Release Clause'].map(standardize)
fifa[['Value', 'Wage', 'Release Clause']]
```

Removing the newline character from the Hits column

```python
fifa.Hits = fifa.Hits.str.replace('\n', '')
```

```python
fifa.Hits
```

Removing the star character from SM and IR column

```python
fifa.SM = fifa.SM.str.replace('★', '')
fifa.IR = fifa.IR.str.replace('★', '')
fifa[['SM', 'IR']]
```

Spliting the Positions column into 3 new columns

```python
fifa[['Position 1', 'Position 2', 'Position 3']] = fifa.Positions.str.split(" ", expand=True)
fifa[['Position 1', 'Position 2', 'Position 3']]
```

Separating the Team & Contract column into 2 different columns

```python
fifa['Team & Contract'] = fifa['Team & Contract'].str.lstrip('\n').str.rstrip('\n')
fifa[['Team', 'Contract']] = fifa['Team & Contract'].str.split('\n', expand=True)
```

Removing columns with not interesting data

```python
fifa = fifa.drop(['photoUrl'], axis='columns')
fifa = fifa.drop(['playerUrl'], axis='columns')
```

Removing duplicated Columns

```python
fifa = fifa.drop(['Joined'], axis='columns')
fifa = fifa.drop(['Team & Contract'], axis='columns')
fifa = fifa.drop(['Positions'], axis='columns')
```
